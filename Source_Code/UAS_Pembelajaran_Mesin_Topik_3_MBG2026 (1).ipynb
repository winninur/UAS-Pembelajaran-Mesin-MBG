```python
# Define a function to clean text
def clean_text(text):
    text = text.lower() # Lowercase the text
    text = re.sub(r'http\S+', '', text) # Remove URLs
    text = re.sub(r'@\w+', '', text) # Remove mentions
    text = re.sub(r'#\w+', '', text) # Remove hashtags
    text = text.translate(str.maketrans('', '', string.punctuation)) # Remove punctuation
    text = re.sub(r'\d+', '', text) # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra spaces
    # Hapus emoji
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)

    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    return text


# Apply the cleaning function to the 'full_text' column
df['cleaned_text'] = df['full_text'].apply(clean_text)

print("DataFrame with cleaned text:")
display(df[['full_text', 'cleaned_text']].head())
```

LDA (Topik Modelling)

```python
!pip install gensim
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import re
import numpy as np
import pandas as pd
import gensim
import nltk
from nltk.stem.snowball import SnowballStemmer
from gensim.corpora import Dictionary, MmCorpus
from gensim.models import ldamodel
from sklearn.cluster import KMeans
from sklearn.manifold import MDS
from sklearn.decomposition import PCA
```

```python
def load_dataset(filename):
    file = open(filename, 'r')
    acc_names = []
    tweets = []
    for line in file:
        line = line.strip()
        parts = line.split('###')
        acc_names.append(parts[0])
        tweets.append(parts[1])
    return acc_names, tweets
```

```python
# config variables
num_topics=4
```

```python
!pip install Sastrawi
import nltk
nltk.download('stopwords')
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
stopwords = nltk.corpus.stopwords.words('indonesian')

def preprocess(text):
    # tokenizing and lowercasing
    tokens = [word.lower() for word in text.split()]
    filtered_tokens = []
    # buat yang bukan terdiri dari alfabet, dan merupakan stopword
    for token in tokens:
        if re.search('[a-zA-Z]', token) and (token not in stopwords):
            filtered_tokens.append(token)
    # lakukan stemming dengan Sastrawi stemmer
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
```

```python
# Kita load dokumen twitter, dan lakukan preprocessing terhadap tweet yang sudah di-load
acc_names, tweets = load_dataset("lda_input.txt")
print(acc_names)
print(tweets)
# Lakukan pre-process untuk setiap tweet pada koleksi "tweets" kita
# Gunakan List Comprehension untuk mempermudah hidup kita
tweets = [preprocess(tweet) for tweet in tweets]
print(tweets)
```

```python
from gensim.corpora import Dictionary

# Membuat Dictionary
dictionary = Dictionary(tweets)

print("Jumlah kata sebelum filtering :", len(dictionary))

# Filter kata
dictionary.filter_extremes(
    no_below=5,      # kata muncul minimal di 5 dokumen
    no_above=0.5     # buang kata yang muncul di >50% dokumen
)

print("Jumlah kata setelah filtering :", len(dictionary))

# Membuat Corpus (Bag of Words)
corpus = [dictionary.doc2bow(tweet) for tweet in tweets]

print(corpus[:5])
```

```python
# Run the LDA !
lda = ldamodel.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, random_state=1, iterations=10000)
```

```python
# tampilkan topic matrix
topics_matrix = lda.show_topics(formatted=False)
for topic_no, topic_words in topics_matrix:
    print ('topic number: {}'.format(topic_no))
    # default: top-10 kata yang paling tinggi probabilitasnya
    for word, prob in topic_words:
        print (word, prob)
```

```python
# bentuk terlebih dahulu vektor dokumen/tweet
# vektor tweet/dokumen = vektor probabilitas terhadap masing-masing topik
tweet_vectors = []
for i, tweet_bow in enumerate(corpus): # Iterate over the corpus (bag-of-words representation)
    # Get topic probabilities for the current document.
    # Setting minimum_probability=0.0 ensures all topics are considered.
    doc_topic_distribution = lda.get_document_topics(tweet_bow, minimum_probability=0.0)

    # Initialize a vector for the current tweet with zeros for all topics
    probs_vector = [0.0] * num_topics

    # Fill the probs_vector with the actual probabilities
    for topic_id, prob in doc_topic_distribution:
        if topic_id < num_topics: # Ensure topic_id is within expected bounds
            probs_vector[topic_id] = prob
    tweet_vectors.append(probs_vector)

tweet_vectors = np.array(tweet_vectors)
# kita set banyaknya cluster = banyaknya topik
num_clusters = num_topics
# gunakan algoritma K-Means, dan lakukan clustering !
km = KMeans(n_clusters=num_clusters, random_state=42, n_init='auto')
km.fit(tweet_vectors)
# jika kita ingin melihat indeks cluster untuk setiap tweet/dokumen
clusters = km.labels_.tolist()
print(clusters)
```

```python
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
cluster_names = {}
for i in range(num_clusters):
    print ("cluster %d words:" % i)

    # ambil 2 topik major untuk setiap cluster
    topic_words = []
    for ind in order_centroids[i, :1]:
            topic_words += [dictionary.get(word_id) for (word_id, prob) in lda.get_topic_terms(ind, topn=2)]

    cluster_names[i] = ','.join(topic_words)
    print (cluster_names[i])
```

Clustering FTC

Clustering menggunakan Frequent Term Based Clustering (FTC). Kandidat cluster dibentuk dari frequent term/term-set, lalu dipilih iteratif memakai nilai Entropy Overlap (EO) minimum

```python
import pandas as pd

df = pd.read_csv("/content/cleaned_text (1).csv")
```

```python
from itertools import combinations
from collections import Counter
from typing import Sequence
import numpy as np
import pandas as pd

FTC_MIN_SUPPORT = 8
FTC_MAX_K = 2
FTC_MAX_TERMS = 120


def generate_frequent_termsets(
    token_lists: Sequence[Sequence[str]],
    min_support: int = 8,
    max_k: int = 2,
    max_terms: int = 120,
) -> dict[tuple[str, ...], set[int]]:
    doc_terms = [set(tokens) for tokens in token_lists]
    doc_freq = Counter()

    for terms in doc_terms:
        doc_freq.update(terms)

    frequent_terms = [term for term, count in doc_freq.most_common(max_terms) if count >= min_support]
    frequent_term_set = set(frequent_terms)
    candidate_counts = Counter()

    for terms in doc_terms:
        filtered_terms = sorted(terms & frequent_term_set)
        for k in range(1, max_k + 1):
            if len(filtered_terms) < k:
                continue
            for termset in combinations(filtered_terms, k):
                candidate_counts[termset] += 1

    candidates = {
        termset: {
            doc_id
            for doc_id, terms in enumerate(doc_terms)
            if set(termset).issubset(terms)
        }
        for termset, support in candidate_counts.items()
        if support >= min_support
    }

    return candidates


def entropy_overlap(candidate_docs: set[int], doc_candidate_frequency: Counter[int]) -> float:
    value = 0.0
    for doc_id in candidate_docs:
        freq = doc_candidate_frequency[doc_id]
        if freq > 1:
            value += np.log(freq) / freq
    return float(value)


def ftc_clustering(
    token_lists: Sequence[Sequence[str]],
    min_support: int = 8,
    max_k: int = 2,
    max_terms: int = 120,
) -> tuple[np.ndarray, pd.DataFrame]:
    candidates = generate_frequent_termsets(
        token_lists,
        min_support=min_support,
        max_k=max_k,
        max_terms=max_terms,
    )

    labels = np.full(len(token_lists), -1, dtype=int)
    remaining_docs = set(range(len(token_lists)))
    selected_rows = []
    cluster_id = 0
    iteration = 1

    while candidates and remaining_docs:
        active_candidates = {
            termset: docs & remaining_docs
            for termset, docs in candidates.items()
        }
        active_candidates = {
            termset: docs
            for termset, docs in active_candidates.items()
            if len(docs) >= min_support
        }

        if not active_candidates:
            break

        doc_candidate_frequency = Counter()
        for docs in active_candidates.values():
            doc_candidate_frequency.update(docs)

        scored_candidates = []
        for termset, docs in active_candidates.items():
            eo_value = entropy_overlap(docs, doc_candidate_frequency)
            scored_candidates.append((eo_value, -len(docs), -len(termset), termset, docs))

        eo_value, neg_support, neg_len, best_termset, best_docs = min(scored_candidates)
        support = len(best_docs)

        labels[list(best_docs)] = cluster_id
        selected_rows.append(
            {
                "iteration": iteration,
                "cluster_id": cluster_id,
                "frequent_termset": ", ".join(best_termset),
                "support": support,
                "entropy_overlap": round(eo_value, 6),
                "n_remaining_docs_after": len(remaining_docs - best_docs),
            }
        )

        remaining_docs -= best_docs
        cluster_id += 1
        iteration += 1

    if remaining_docs:
        residual_cluster_id = cluster_id
        labels[list(remaining_docs)] = residual_cluster_id
        selected_rows.append(
            {
                "iteration": iteration,
                "cluster_id": residual_cluster_id,
                "frequent_termset": "RESIDUAL_UNASSIGNED",
                "support": len(remaining_docs),
                "entropy_overlap": 0.0,
                "n_remaining_docs_after": 0,
            }
        )

    return labels, pd.DataFrame(selected_rows)


def topic_label(topic_words_df: pd.DataFrame, topic_id: int, n_words: int = 5) -> str:
    words = (
        topic_words_df.loc[topic_words_df["topic_id"] == topic_id]
        .sort_values("rank")
        .head(n_words)["term"]
        .tolist()
    )
    return ", ".join(words)


def make_cluster_summary(
    comments_df: pd.DataFrame,
    topic_words_df: pd.DataFrame,
    token_lists: Sequence[Sequence[str]],
    ftc_selected_clusters: pd.DataFrame,
) -> pd.DataFrame:
    rows = []
    termset_lookup = dict(
        zip(ftc_selected_clusters["cluster_id"], ftc_selected_clusters["frequent_termset"])
    )
    eo_lookup = dict(
        zip(ftc_selected_clusters["cluster_id"], ftc_selected_clusters["entropy_overlap"])
    )

    for cluster_id, group in comments_df.groupby("cluster_id"):
        dominant_topic = int(group["topic_id"].mode().iloc[0])
        token_counter = Counter()

        for index in group.index:
            token_counter.update(token_lists[index])

        rows.append(
            {
                "cluster_id": int(cluster_id),
                "n_comments": int(len(group)),
                "ftc_termset": termset_lookup.get(int(cluster_id), ""),
                "entropy_overlap": eo_lookup.get(int(cluster_id), 0.0),
                "dominant_topic": dominant_topic,
                "dominant_topic_terms": topic_label(topic_words_df, dominant_topic),
                "cluster_top_terms": ", ".join(term for term, _ in token_counter.most_common(10)),
                "example_comment": str(group[TEXT_COL].iloc[0]),
            }
        )

    return pd.DataFrame(rows).sort_values("cluster_id").reset_index(drop=True)
```

```python
token_lists = tweets
lda_model = lda
N_TOPICS = num_topics

# Create topic_words DataFrame from LDA model output
topic_words_data = []
for topic_id, word_probs in lda_model.show_topics(formatted=False, num_topics=N_TOPICS):
    for i, (word, prob) in enumerate(word_probs):
        topic_words_data.append({"topic_id": topic_id, "term": word, "rank": i, "probability": prob})
topic_words = pd.DataFrame(topic_words_data)

cluster_labels, ftc_selected_clusters = ftc_clustering(
    token_lists,
    min_support=FTC_MIN_SUPPORT,
    max_k=FTC_MAX_K,
    max_terms=FTC_MAX_TERMS,
)

# The 'tokens' column does not exist in the current df, so remove the drop operation.
result_df = df.copy()
# Use already computed all_dominant_topics for topic_id
result_df["topic_id"] = all_dominant_topics
# Use tweet_vectors for topic_confidence
result_df["topic_confidence"] = tweet_vectors.max(axis=1).round(4)
result_df["cluster_id"] = cluster_labels

for topic_id in range(N_TOPICS):
    # Use tweet_vectors for individual topic probabilities
    result_df[f"topic_{topic_id}_prob"] = tweet_vectors[:, topic_id].round(4)

# Define TEXT_COL for the make_cluster_summary function
TEXT_COL = 'cleaned_text'

cluster_summary = make_cluster_summary(
    result_df,
    topic_words,
    token_lists,
    ftc_selected_clusters,
)

print("Jumlah cluster FTC:", result_df["cluster_id"].nunique())
print("Min support FTC:", FTC_MIN_SUPPORT)
print("Max k termset:", FTC_MAX_K)
cluster_summary
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load data
edges = pd.read_csv("edges.csv")

# Membuat graph
G = nx.from_pandas_edgelist(
    edges,
    source="Source",
    target="Target",
    create_using=nx.Graph()
)

print("Jumlah Node :", G.number_of_nodes())
print("Jumlah Edge :", G.number_of_edges())
```

Hitung SNA

```python
import pandas as pd

degree = nx.degree_centrality(G)
betweenness = nx.betweenness_centrality(G)
closeness = nx.closeness_centrality(G)
eigenvector = nx.eigenvector_centrality(G, max_iter=1000)

hasil_sna = pd.DataFrame({
    "Node": list(degree.keys()),
    "Degree": list(degree.values()),
    "Betweenness": [betweenness[n] for n in degree.keys()],
    "Closeness": [closeness[n] for n in degree.keys()],
    "Eigenvector": [eigenvector[n] for n in degree.keys()]
})

hasil_sna = hasil_sna.sort_values(
    by="Degree",
    ascending=False
)
hasil_sna.to_excel(
    "hasil_SNA.xlsx",
    index=False
)

print("Berhasil disimpan")

hasil_sna.head()
```

Visualisasi Network

```python
plt.figure(figsize=(12,10))

pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_nodes(
    G,
    pos,
    node_size=40
)

nx.draw_networkx_edges(
    G,
    pos,
    alpha=0.3
)

plt.title("Social Network Analysis")
plt.axis("off")
plt.show()
```
